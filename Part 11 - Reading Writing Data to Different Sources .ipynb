```python
filt = df['Country'] == 'Turkey'
turkey_df = df.loc[filt]
turkey_df.head(5)
```

```python
turkey_df.to_csv('datas/survey_results_public_turkey.csv')
```

```python
turkey_df.to_csv('datas/survey_results_public_turkey.tsv', sep='\t')
```
